# Seaborn
O Seaborn é uma biblioteca de visualização de dados do Python baseado no Matplotlib. Ele provê uma interface de alto nível para construção de gráficos estatísticos atrativos e informativos. Site Seaborn org

```python
import numpy as np
import pandas as pd
import seaborn as sns

# Definir tema (estilo dark pros gráficos)
sns.set_theme(style='darkgrid')
```

```python
# pegando uma base de dados do proprio seaborn - tips = sobre restaurantes
Base_dados = sns.load_dataset('tips')

# verificando:
Base_dados.head()
```

```python
# renomeando as colunas:
Base_dados.rename(columns={
    'total_bill' : 'Total Conta',
    'tip' : 'Gorjeta',
    'sex' : 'Sexo',
    'smoker' : 'Fumante',
    'day' : 'Dia semana',
    'time' : 'Período',
    'size' : 'Pessoas na mesa'},
    inplace=True)
Base_dados.head()
```

# Gráfico relplot

```python
# Gráfico relplot
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_dados );
```

```python
# Gráfico relplot (SEPARANDO POR SEXO --> HUE)
# Passando outro parametro como classe no gráfico
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_dados, hue='Sexo' );
```

```python
# Gráfico de Linha com 2 Eixos (MIDANDO P GRÁFICO TIPO LINHA --> KIND LINE)
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_dados, kind='line' );
```

```python
# Gráfico de Linha com 2 Eixos (SEPARANDO QUEM É FUMANTE E QUEM NÃO É --> HUE)
sns.relplot( x='Total Conta', y='Gorjeta', data=Base_dados, 
            kind='line', hue='Fumante' );
```

# Gráfico histplot

```python
sns.histplot( data=Base_dados, x='Total Conta')
# A MAIORIA DAS GORJETAS SE CONCENTRAM ENTRE 10 E 15 REAIS
```

```python
# SEPARANDO AS INFOS ENTRE FUMANTES E NÃO FUMANTES
sns.histplot( data=Base_dados, x='Total Conta', hue='Fumante');
# OS NÃO FUMANTES DÃO MAIORES GORJETAS
```

# Gráfico barplot

```python
#SEPARANDO EM SEXOS E FUMANTES
sns.barplot( data=Base_dados, x='Sexo', y='Total Conta', hue='Fumante');
```

# Gráfico pairplot

```python
sns.pairplot( Base_dados );
```

```python
sns.pairplot( Base_dados, hue='Sexo' );
#MESMA PLOTAGEM ANTERIOR, SEPARANDO POR SEXO
```

```python
sns.pairplot( Base_dados, kind='kde' );
```

# Gráfico boxplot

```python
sns.boxplot( data=Base_dados, x='Dia semana', y='Total Conta');
```

```python
sns.boxplot( data=Base_dados, x='Dia semana', y='Total Conta', hue='Sexo');
```

# Gráfico violinplot

```python
sns.violinplot( data=Base_dados, x='Dia semana', y='Total Conta' );
```
